```python
#set release date in perfect fromat
df_tracks.set_index("release_date",inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks[["artists"]].iloc[23]
```

```python
# convert duration column unit milisecond->second
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms',inplace = True,axis = 1)
```

```python
df_tracks.duration.head()
```

```python
#five top durated songs
listeners = df_tracks.sort_values("duration",ascending =False).head(5)
listeners
```

```python
#correlation map
corr_df = df_tracks.drop(['key','mode','explicit'],axis=1).corr(method = 'pearson')
plt.figure(figsize=(14,6))
heatmap = sb.heatmap(corr_df,annot = True,fmt=".1g",vmin = -1,center=0,cmap = 'inferno',linewidths =1,linecolor='Black')
heatmap.set_title("Correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation =90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sb.regplot(data = sample_df, y = 'loudness', x = "energy",scatter_kws={"color": "c"}, line_kws={"color": "black"}).set(title = "Loudness vs Energy Correlation")
```

```python
plt.figure(figsize=(10,6))
sb.regplot(data = sample_df, y = 'popularity', x = "acousticness",scatter_kws={"color": "g"}, line_kws={"color": "black"}).set(title = "Popularity vs Acousticness Correlation")
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
years
```

```python
sb.set_style('whitegrid')
sb.distplot(years,kde = True, color ='c', bins = 30).set(title = "Number of songs per year")
```

```python
total_dr = df_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sb.barplot(x=years,y = total_dr,ax=ax,errwidth =False).set(title = "Year vs Duration")
plt.xticks(rotation =90)
plt.show()
```

```python
total_dr = df_tracks.duration
sb.set_style(style ="whitegrid")
fig_dims = (10,5)
fig, ax = plt.subplots(figsize = fig_dims)
fig =sb.lineplot(x= years,y = total_dr,ax=ax).set(title = "Year vs Duration")
plt.xticks(rotation =60)
plt.show()
```

# insights

* As the Loudness of the song increases, so does its energy.
* Number of song releases have increased with time.
* The popularity column is unbalanced. The graph is left - skewed. Songs with popularity near zero has major count.
* Most of the songs have low speechiness i.e. tracks with music only and no - speech. More Popular songs contain somewhat speechiness but many of the songs are explicit.
* Most of the speechiness Observations are in the range 0.0 to 0.6. Some of the values are in the range 0.8 to 1.0. A non - normal distribution.
